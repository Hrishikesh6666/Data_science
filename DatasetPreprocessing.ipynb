# Dealing With Null Value

```python
df.isnull().sum()
```

```python
df['Rating'].describe()
```

```python
from sklearn.impute import SimpleImputer  # For handaling the null value
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
# Reshaping the data for Imputation

rating =  np.reshape(df['Rating'].values, (10841,1))
imputer.fit(rating)
rating = imputer.transform(rating)
df['Rating'] = rating
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)  # removing the unrequired small unrequired data
```

### Converting size into Number(MB)

```python
# Handeling the "Varies with device"

size=[]
for i in df['Size']:
    if i == 'Varies with device':
        size.append(10)
    elif i[-1] == 'M':
        size.append(float(i[:-1]))
    elif i[-1] == 'k':
        size.append(float(i[:-1])/1024)
    else:
        size.append(i)
df['Size'] = size
```

```python
df['Size'].describe()
```

```python
# AVERAGE SIZE OF APPLICATION

print("Average Size : ",round(df['Size'].mean(),2))
print("Maximum Size : ",df['Size'].max())
print("Minimum Size : ",df['Size'].min())
```

### Convert the USE $ TO another curreny

```python
price=[]

for i in df['Price']:
  if('$' in i):
    price.append(float(i[1:])*85)
  else:
    price.append(0)
df['Price'] = price
```
